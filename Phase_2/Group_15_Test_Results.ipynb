```python
import import_ipynb
import warnings
warnings.filterwarnings('ignore')
from sklearn import metrics
from sklearn.metrics import accuracy_score,recall_score
import matplotlib.pyplot as plt
import pickle
```

```python
from Group_15_Linear_Classification import lr,random_model,X_test,y_test
```

### Random Model - Baseline

```python
rd_pred = random_model.predict(X_test)
```

```python
fpr,tpr,_ = metrics.roc_curve(y_test,rd_pred)
auc = round(metrics.roc_auc_score(y_test,rd_pred),4)
```

### Linear Regression

```python
lr_pred,MSE = lr.predict_lr(X_test)
```

```python
fpr_l,tpr_l,_ = metrics.roc_curve(y_test,lr_pred)
auc_l = round(metrics.roc_auc_score(y_test,rd_pred) , 4)
```

### Logistic Regression

```python
from Group_15_Logistic_Classification import lr_full
```

```python
log_pred = lr_full.predict(X_test)
```

```python
fpr_lg,tpr_lg,_ = metrics.roc_curve(y_test,log_pred)
auc_log = round(metrics.roc_auc_score(y_test,log_pred) , 4)
```

### Random Forest

```python
rf= pickle.load(open('randomforest_model.sav', 'rb'))
```

```python
rf_preds = rf.predict(X_test)
```

```python
fpr_rf,tpr_rf,_ = metrics.roc_curve(y_test,rf_preds)
auc_rf = round(metrics.roc_auc_score(y_test,rf_preds) , 4)
```

### K - Nearest Neighbors

```python
knn = pickle.load(open('KNN_model.sav', 'rb'))
```

```python
knn_preds = knn.predict(X_test)
```

```python
fpr_knn,tpr_knn,_ = metrics.roc_curve(y_test,knn_preds)
auc_knn = round(metrics.roc_auc_score(y_test,knn_preds) , 4)
```

### ROC Curve

```python
plt.figure(figsize=(6,4))
plt.plot(fpr,tpr,label = "Random AUC = "+str(auc))
plt.plot(fpr_l,tpr_l,label = "Linear AUC = "+str(auc_l))
plt.plot(fpr_lg,tpr_lg,label = "Logistic AUC = "+str(auc_log))
plt.plot(fpr_rf,tpr_rf,label = "Random Forest AUC = "+str(auc_rf))
plt.plot(fpr_knn,tpr_knn, label = "KNN AUC = "+str(auc_knn))
plt.legend(loc = 4)
plt.show()
```

### Accuracies

```python
print("The accuracy of the Random model is", round(accuracy_score(y_test, rd_pred),2)*100)
print("The accuracy of the Linear Regression is", round(accuracy_score(y_test, lr_pred),2)*100)
print("The accuracy of the Logistic Classification is", round(accuracy_score(y_test, log_pred),2)*100)
print("The accuracy of the Random Forest is", round(accuracy_score(y_test, rf_preds),2)*100)
print("The accuracy of the KNN is", round(accuracy_score(y_test, knn_preds),2)*100)
```

### Comparing Re-call Values

```python
print("The recall of the Random model is", round(recall_score(y_test, rd_pred),2))
print("The recall of the Linear Regression is", round(recall_score(y_test, lr_pred),2))
print("The recall of the Logistic Classification is", round(recall_score(y_test, log_pred),2))
print("The recall of the Random Forest is", round(recall_score(y_test, rf_preds),2))
print("The recall of the KNN is", round(recall_score(y_test, knn_preds),2))
```
